```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
```

# Compute the correlation matrix

```python
numerical_cols = df.select_dtypes(include=[np.number])
corr_matrix = df.corr()
```

```python
corr_matrix
```

```python
plt.figure(figsize=(12, 10))
sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap='coolwarm')
plt.title('Correlation Matrix of Variables')
plt.show()
```

```python
sns.pairplot(numerical_cols[['PropertyID', 'YearBuilt', 'SaleYr', 'PoolArea', 'BsmntSqFtage',
       'PropertyFrontage', 'OpenPorchSF','PropPrice']], height=5)
plt.show()
```

# PropPrice distribution

```python
plt.figure(figsize=(8, 6))
sns.histplot(data=df, x='PropPrice', kde=True)
plt.title('Distribution of PropPrice')
plt.show()
```

# Violin plots

```python
plt.figure(figsize=(10, 6))
sns.violinplot(data=df, y='PropPrice')
plt.ylabel('PropPrice')
plt.title('Violin Plot of PropPrice')
plt.show()
```

# Count Plot

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='Neighborhood_BrDale', data=df)
plt.xticks(rotation=90)
plt.title('Properties per Neighborhood_BrDale')
plt.xlabel('Neighborhood_BrDale')
plt.ylabel('Count')
plt.show()
```

# Analyze the relationship between  feature and target.

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='GrLivArea', y='PropPrice', data=df)
plt.title('Living Area vs. Property Price')
plt.xlabel('Ground Living Area (sq ft)')
plt.ylabel('Property Price')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='PropertyFrontage', y='PropPrice', data=df)
plt.title('PropertyFrontage vs. Property Price')
plt.xlabel('PropertyFrontage')
plt.ylabel('Property Price')
plt.show()
```
